```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import pandas as pd
import datetime
import time


# Initialize the Selenium WebDriver (e.g., Chrome)
driver = webdriver.Chrome()

# URL of the initial page to scrape
base_url = 'https://www.hkstp.org/zh-hk/directory'

# Open the initial page
driver.get(base_url)

# Function to wait until the page is fully loaded
def wait_for_element(driver):
    WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.XPATH, "//template[@x-for='result in results']"))
    )

# Function to extract the table from the current page
def extract_table_from_page(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table')
    
    # Convert the table to a pandas DataFrame
    df = pd.read_html(str(table))[0]
    return df

# Function to extract strings from <div class="card">
def extract_strings_from_cards(driver):
    print("loading strings from cards")
    #html = driver.page_source
    #soup = BeautifulSoup(html, 'html.parser')
    #cards = soup.find_all('div', class_='card-body')
    
    #strings = [(card.get_text(strip=True),"\t") for card in cards]
    #return strings

    # Find all the card elements
    cards = driver.find_elements(By.CLASS_NAME, 'card')

    # Initialize the arrays to store text and names
    text_array = []
    name_array = []
    
    # Iterate through each card and extract text and name
    for card in cards:
        #text = card.find_element(By.XPATH, '//div[@class="tags"]/a[@class="tag"]').text
        #name = card.find_element(By.XPATH, '//div[@class="card-body"]/a[@class="name"]').text
        text = card.find_element(By.CLASS_NAME, 'tag').text
        name = card.find_element(By.CLASS_NAME, 'name').text
        text_array.append(text)
        name_array.append(name)
    return text_array, name_array
    
# Initialize an empty DataFrame to store all tables
all_tables = pd.DataFrame()
# Initialize an empty list to store strings from cards
all_text = []
all_name = []
current_page = 1
time.sleep(10)
while True:
    #print(f'Scraping {driver.current_url}')
    # Wait for the page to load completely
    #wait_for_element(driver)
    
    # Extract the table from the current page
    #table_df = extract_table_from_page(driver)
    #all_tables = pd.concat([all_tables, table_df])
    
    # Extract strings from <div class="card">
    #strings = extract_strings_from_cards(driver)
    #all_strings.extend(strings)
    
    try:
        # Wait until the next button is clickable and click it
        print("Start wait")
        try:
            WebDriverWait(driver, 30).until(EC.invisibility_of_element_located((By.CLASS_NAME, 'column-4-photo-card-listing.component.is-logo.loading')))
        except TimeoutException:
            print("Loading element did not disappear in time")
            break
        print("wait yuen")
        
        print("You are now in page ", current_page)
        text_array, name_array = extract_strings_from_cards(driver)
        all_text.extend(text_array)
        all_name.extend(name_array)
        
        #next_button = WebDriverWait(driver, 10).until(
        #    EC.element_to_be_clickable((By.CLASS_NAME, 'next'))
        #)
        next_button = driver.find_element(By.XPATH, '//div[@class="pagination"]/button[@class="next"]')
        driver.execute_script("arguments[0].click();", next_button)
        ##BUG HERE
        ##Cannot auto top here
        #next_button.click()
        #next_button = driver.find_element(By.XPATH, '//button[@class="next"]')
        #actions = ActionChains(driver)
        #actions.move_to_element(next_button).click().perform()
        current_page = current_page + 1
        mycount = 1/(87-current_page)
        # Wait for the page to load completely after clicking the next button
        #wait_for_page_load(driver)
        
    except Exception as e:
        print(f"No more pages to scrape: {e}")
        break

# Close the WebDriver
driver.quit()

# Save the collected tables to a CSV file
all_tables.to_csv('scraped_tables.csv', index=False)

# Save the collected strings to a text file
with open('card_strings.txt', 'w') as f:
    for string in all_strings:
        f.write(string + '\n')
```

```python
def combine_arrays_into_table(array1, array2):
    # Ensure both arrays have the same length
    if len(array1) != len(array2):
        raise ValueError("Both arrays must have the same length")

    # Create a DataFrame from the two arrays
    data = {
        'Column1': array1,
        'Column2': array2
    }
    df = pd.DataFrame(data)
    df.replace("", pd.NA, inplace=True)
    df.dropna(inplace=True)
    return df

def save_table_to_csv(df, filename):
    # Drop rows with any NaN values

    
    # Save the DataFrame to a CSV file
    df.to_csv(filename, index=False, encoding='utf-8')

table = combine_arrays_into_table(all_text, all_name)

# Save the table to a CSV file
filename = 'combined_table2.csv'
save_table_to_csv(table, filename)

print(f"Table saved to {filename}")
```
